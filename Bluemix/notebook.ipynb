```scala
import org.apache.spark.mllib.clustering.{KMeans, KMeansModel}
import org.apache.spark.mllib.linalg.Vectors
import scala.util.Try
```

```scala
val rdd = sc.textFile("swift://kmeans.spark/spark-dataset100")
val kmeans100 = rdd.map(line => Try(Vectors.parse(line))).filter(_.isSuccess).map(_.get)
```

```scala
val numClusters = 2
val numIterations = 20
val t0 = System.nanoTime()
val clusters = KMeans.train(kmeans100, numClusters, numIterations)
val WSSSE = clusters.computeCost(kmeans100)
    println("Within Set Sum of Squared Errors = " + WSSSE)
val t1 = System.nanoTime()

println("Elapsed time: " + (t1 - t0) + "ns")
```
